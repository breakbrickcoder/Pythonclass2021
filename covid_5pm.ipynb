```python
from bs4 import BeautifulSoup
import requests
url = BeautifulSoup('https://www.worldometers.info/coronavirus/','html.parser')
response = requests.get(url)
response
```

```python
data = response.text
data
```

```python
soup = BeautifulSoup(data,'lxml')
soup
```

```python
soup = soup.table
soup
```

```python
tag = soup.find_all('tr')
tag
```

```python
a = []
for i in tag:
  x = i.get_text()
  a.append(x)

a
```

```python
output = []
for i in a:
   x = i.split('\n')[1:-1]
   output.append(x)

output
```

```python
del output[1:9]
```

```python
del output[-9::1]
```

```python
output
```

```python
import csv
f = open('covid.csv','w')
write = csv.writer(f)
for i in output:
  write.writerow(i)

f.close()
```
