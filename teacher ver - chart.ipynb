```python
s = str()
b = open('data.csv','w')
b.write("Name, Age, Phone, Address, Salary\n")
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    phone = int(input("Enter phone = "))
    address = input("Enter address = ")
    salary = int(input("Enter salary = "))
    info = f"{name},{age},{phone},{address},{salary}\n"
    s = s+info
    print(i,s)
    
b.write(s)
b.close()
```

```python
# append
from csv import writer
list = ['9', 'nima', 'pokhara', '45', '949489458', '48000']
with open('data.csv','a') as f_object:
    writer_object  = writer(f_object)
    writer_object.writerow(list)
    f_object.close()
```

```python
from csv import DictWriter
label = ['Sn', 'Name', 'Address', 'Age', 'Phone', 'Salary']
dict = {'Sn':'10','Name':'Arjun','Address':'lukla','Age':'43','Phone':'9484854433','Salary':'80000'}

with open('data.csv','a') as f_object:
    dictwriter_object = DictWriter(f_object,label)
    dictwriter_object.writerow(dict)
    f_object.close()
```
